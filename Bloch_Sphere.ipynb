# What is Bloch Sphere

representing a state that is **a combination of both 0 or 1**

## Knowing that r = 1 (radius of a unit sphere is 1)




**With spherical coordinates defined as such:**



###
$$
\begin{aligned}
x &= r \sin \theta \cos \phi \\
y &= r \sin \theta \sin \phi \\
z &= r \cos \theta
\end{aligned}
$$

```python
import numpy as np 
```

```python
def tf(theta, pi):
    x = np.sin(theta)* np.cos(pi)
    y = np.sin(theta) * np.sin(pi)
    z = np.cos(theta)
    return x,y,z
```

### Visualization with Qiskit

```python
from qiskit.visualization import plot_bloch_vector
```

#### 
$$
|0\rangle = \cos \frac{0}{2} |0\rangle + e^{i0} \sin \frac{0}{2} |1\rangle
$$

```python
theta = 0
pi = 0 

x,y,z = tf(theta,pi)

plot_bloch_vector([x,y,z])
```

#### 
$$
|1\rangle = \cos \frac{\pi}{2} |0\rangle + e^{i0} \sin \frac{\pi}{2} |1\rangle
$$

```python
theta = np.pi
pi = 0 

x,y,z = tf(theta,pi)

plot_bloch_vector([x,y,z])
```

#### 
$$
|1\rangle = \cos \frac{\pi/2}{2} |0\rangle + e^{i\pi} \sin \frac{\pi/2}{2} |1\rangle = \frac{|0\rangle - |1\rangle}{\sqrt{2}}
$$

```python
theta = np.pi/2
phi = np.pi

x,y,z = tf(theta,phi)

plot_bloch_vector([x,y,z])
```
